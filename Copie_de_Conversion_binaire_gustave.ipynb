```python
import math
import time  # Import the time module for pausing

# Lecture de l'entrée
N = int(input("Entrez un nombre entier : "))

# Initialisation
quotient = N
resultat = []
indice = 0

print(f"\nConversion de {N} en binaire (mode pas à pas) :")
time.sleep(1) # Pause for 1 second

# Conversion en binaire
while quotient != 0:
    reste = quotient % 2
    quotient = math.floor(quotient / 2)
    resultat.append(reste)
    print(f"  {N if indice == 0 else int(N / (2**indice))} // 2 = {quotient}, Reste = {reste}")
    time.sleep(1) # Pause for 1 second
    indice += 1


# Affichage du résultat en inversant l'ordre
print("\nRésultat en binaire (ordre inversé) : ", end="")
for i in range(len(resultat) - 1, -1, -1):
    print(resultat[i], end="")

print("\n") # Saut de ligne final
```
