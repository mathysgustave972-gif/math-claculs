```python
# lire N
N = int(input('entre N :'))
quotient = N // 2
reste = N % 2
indice = 0
resultat= []
while quotient != 0:
  resultat.append (reste)
  reste = quotient % 2
  quotient = quotient //2
  indice = indice + 1
resultat.append (1)

#print(resultat)
#print(indicee)

while indice >= 0:
  print(resultat[indice], end='')
  indice -= 1
```
